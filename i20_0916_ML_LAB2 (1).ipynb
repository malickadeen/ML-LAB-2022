**OBJECT ORIENTED PROGRAMING**

**Protected Members:**

```python
class person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

obj3 = person("Osama", 102)
obj3._name
obj3._name = "Shayyan"
obj3._name
```

**Private Members:**

```python
class person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

obj3 = person("Osama", 102)
obj3.__name
obj3.__name = "Shayyan"
obj3.__name
```

**Access modifiers**

```python
class Members:
    member1 = None  #public member
    _member2 = None  #protected member
    __member3 = None   #private member
    
    def __init__(self, member1, _member2, __member3):
        self.member1 = member1
        self._member2 = _member2
        self.__member3 = __member3

    def displayPublic(self):
        print("Public Data Member: ", self.member1)
    
    def _displayProtected(self):
        print("Protected Data Member: ", self._member2)
    
    def __displayPrivate(self):
        print("Private Data Member: ", self.__member3)
    
    def accessPrivate(self):	  #public function to access private members
        self.__displayPrivate()


obj = Members("Adeen", "Amir", "Ali")
obj.displayPublic()
obj._displayProtected()
obj.accessPrivate()
```

__Length of string more than 2 and first and last element identical:__

```python
L2=["Adeen","Ali","Amir","Akash","Osama","Shayyan","Adeel","a","alia"]
length=[]
identical=[]
for ind, keys in enumerate(L2):
    if(len(L2[ind])>=2):
        length.append(L2[ind])
        
for ind, keys in enumerate(length):
        if(length[ind][0]==length[ind][-1]):
            identical.append(length[ind])
            
print("Number of words whose length is greator than 2 are ")
print(len(length))
print("Number of words whose first and last element is identical are ")
print(len(identical))
```

__Sorting alphabetically and grouping X elements__

```python
L3=["Adeen","Ali","Amir","Akash","Osama","Shayyan","Adeel","Alia","Babu Bhai", "Xavior","Xotic","Xray","Xsmax"]
size=len(L3)
x_sort=[]
for i, val in enumerate(L3):
    for j in range(0,size-i-1):
        if(L3[j][0] > L3[j+1][0]):
            temp = L3[j]
            L3[j]= L3[j + 1]
            L3[j + 1]= temp 
print(L3)

for i, val in enumerate(L3):
    if(L3[i][0]=="X" or L3[i][0]=="X"):
        x_sort.append(L3[i])
print(x_sort)
```

__Sort using second value of tuple__

```python
L4 = [(6,7),(8,1),(1,44),(8,8),(19,76),(5,21),(5,34)]
size = len(L4)
for i,val in enumerate(L4):
    for j in range(0, size-i-1):
        if (L4[j][1] > L4[j + 1][1]):
            temp = L4[j]
            L4[j]= L4[j + 1]
            L4[j + 1]= temp        
print(L4)
```

__Remove repeating value in a string__

```python
L5=[1,2,2,3,3,4,5,6,7,8,9,9,9]
size=len(L5)
res=[]
for i, val in enumerate(L5):
    if val not in res:
        res.append(val)
print(res)
```

__Merge two lists and sort them__

```python
L6=[1,2,3,4,5,6,7,8,9]
L7=[1,3,7,13,14,15,16]
L8=L6+L7
size=len(L8)
for i in range(size):
    for j in range(0, size-i-1):
        if (L8[j] > L8[j + 1]):
            temp = L8[j]
            L8[j]= L8[j + 1]
            L8[j + 1]= temp 
print(L8)
```
